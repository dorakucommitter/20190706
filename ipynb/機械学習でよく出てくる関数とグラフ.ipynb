# 標準正規分布グラフ

平均μ=0, 分散σ=1の正規分布

$$
    f(x) = \frac{1}{\sqrt{ 2 \pi}} e^{-\frac{x^2}{2}}
$$

````python
import matplotlib.pyplot as plt
import numpy as np
import math
````

````python
# f(x)を自分で書く
f = lambda x: (math.exp(-x**2/2)) / math.sqrt(2*math.pi)

# ベクトルxを [-5.0, ..., 5.0] の区間で作成
n = np.linspace(-5.0, 5.0, 10000)

# f(x)の結果を得る
p = []
for i in range(len(n)):
    p.append(f(n[i]))

# グラフに表示
fig, ax = plt.subplots()
ax.plot(n, p)
ax.set_yscale('linear')
ax.grid()

plt.show()
````

# 活性化関数

## ステップ関数

````python
def step_function(x):
  y = x > 0
  return y.astype(np.int)

x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
````

## シグモイド関数

$$
    f(x) = \frac{1}{1 + e^{-ax}} (a > 0)
$$

````python
x = np.arange(-10, 10, 0.1)
e = math.e

y = 1 / (1 + e**-x)

# グラフに表示
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_yscale('linear')
ax.grid()

plt.show()
````

## ReLU関数

$$
    ReLU(x) =
        \begin{cases}
            x \quad x \geqq 0 \\
            0 \quad x < 0 \\
        \end{cases}
$$

````python
def relu(x):
  return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)

# グラフに表示
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_yscale('linear')
ax.grid()

plt.show()
````

## ハイパボリックタンジェント

$$
    tanh x = \frac{ e^x - e^{-x} }{ e^x + e^{-x} }
$$

````python
x = np.arange(-5, 5, 0.1)
y = np.tanh(x)

# グラフに表示
fig, ax = plt.subplots()
plt.plot(x, y)
ax.set_yscale('linear')
ax.grid()

plt.show()
````

# その他の関数

## ソフトマックス

数値配列の各要素の値に対し、配列の総和を100%としたとき、何%に相当するかを計算してくれる関数。

$$
    f(y_i) = \frac{ e^{y_i} }{\sum_{j}e^{y_i}}
$$

````python
def softmax(a):
  exp_a = np.exp(a)
  sum_exp_a = np.sum(exp_a)
  y = exp_a / sum_exp_a
  return y

x = np.arange(-5, 10, 0.1)
y = softmax(x)

# グラフに表示
fig, ax = plt.subplots()
plt.plot(x, y)
ax.set_yscale('linear')
ax.grid()

plt.show()
````
